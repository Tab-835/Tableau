```python
import pandas as pd
import numpy as np
```

```python
url_jan = "201701-citibike-tripdata.csv"
jan_df = pd.read_csv(url_jan)
jan_df.head()
```

```python
url_feb = "201702-citibike-tripdata.csv"
feb_df = pd.read_csv(url_feb)
feb_df.head()
```

```python
url_mar = "201703-citibike-tripdata.csv"
mar_df = pd.read_csv(url_mar)
mar_df.head()
```

```python
data_df = pd.concat([jan_df, feb_df, mar_df], ignore_index=True)
```

```python
data_df.dropna()
data_df[data_df.Gender!=0]
```

```python
data_df.dtypes
```

```python
data_df.head()
```

```python
new_data_df = data_df.drop(['Stop Time', 'Start Station ID', 'End Station ID', 'Bike ID'], axis=1)
```

```python
new_data_df.head()
```

```python
new_data_df = new_data_df[new_data_df.Gender!=0]
```

```python
new_data_df = new_data_df.replace({'Gender': 1}, {'Gender': 'Male'}, regex=True)
```

```python
new_data_df = new_data_df.replace({'Gender': 2}, {'Gender': 'Female'}, regex=True)
```

```python
new_data_df["age"] = 2019-new_data_df["Birth Year"]
```

```python
new_data_df
```

```python
new_data_df.to_csv("final_datav1.csv", encoding='utf-8', index=False)
```
